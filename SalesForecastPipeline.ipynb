# Churn Prediction Analysis

In this notebook, we analyze customer churn using visualization and model predictions.


## 1. Data Overview

We use a sample dataset to understand customer behavior and predict churn likelihood.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Customer': ['A', 'B', 'C', 'D'],
    'MonthlyCharges': [70, 99, 85, 60],
    'Churn': ['Yes', 'No', 'Yes', 'No']
}
df = pd.DataFrame(data)
```

```python
# Pie Chart for Churn Distribution
churn_counts = df['Churn'].value_counts()
plt.figure(figsize=(4,4))
churn_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
plt.title("Churn Distribution")
plt.ylabel('')
plt.tight_layout()
plt.show()
```

## 2. Monthly Charges Analysis

Let’s visualize the monthly charges across customers.


```python
plt.figure(figsize=(6,4))
plt.bar(df['Customer'], df['MonthlyCharges'], color='skyblue')
plt.title("Monthly Charges per Customer")
plt.xlabel("Customer")
plt.ylabel("Monthly Charges")
plt.grid(True)
plt.tight_layout()
plt.show()
```
